```python
import pandas as pd
import os

df = pd.read_table("./"+ os.environ['ipc'] +"_energy_time_"+ os.environ['node'] +".txt", sep=' ', na_values='NaN')
```

```python
correlate = df.corr(method=os.environ['correlation'])
correlate
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))
heatmap = sb.heatmap(correlate, 
            xticklabels=correlate.columns,
            yticklabels=correlate.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5)

ipc = "" 
if str(os.environ['ipc']) == 'rest':
    plt.title("Energy-Time Correlation for " + os.environ['cpu'] + " Rest " + str(os.environ['node'][0].upper() + os.environ['node'][1:]), fontsize =20)
    #ipc = str(os.environ['ipc'][0].upper() + os.environ['ipc'][1:])
    
if str(os.environ['ipc']) == 'rpc':
    plt.title("Energy-Time Correlation for " + os.environ['cpu'] + " RPC " + str(os.environ['node'][0].upper() + os.environ['node'][1:]), fontsize =20)
    #ipc = os.environ['ipc'].upper()
    
if str(os.environ['ipc']) == 'grpc':    
    plt.title("Energy-Time Correlation for " + os.environ['cpu'] + " gRPC " + str(os.environ['node'][0].upper() + os.environ['node'][1:]), fontsize =20)
    


fig = heatmap.get_figure()
fig.savefig(os.environ['cpu'] + "_energy_time_comparison_"+ os.environ['node'] + "_" + os.environ['ipc'] +".pdf")
```
