```python
import pandas as pd


filepath='grpc.txt' 
energy = pd.read_table(filepath, delim_whitespace=True, header=0, na_values=-1)
energy.head()
```

```python
energy.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

ipc = filepath.split('.')

for column in energy:
    filename = ipc[0] + "_" + energy[column].name + ".pdf"
    newEnergy =energy[column].fillna(method='ffill')
    sns_plot = sns.distplot(newEnergy, hist=True)
    sns_plot.figure.savefig(filename)
    plt.figure()
```
